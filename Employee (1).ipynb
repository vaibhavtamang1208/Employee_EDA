```python
# Bivariate Analysis
# Physical Activity vs dept
```

```python
# Scatter plot
plt.scatter(df['PhysicalActivityHours'], df['Dept'])
plt.show()
```

```python
# Bar chart
mean_activity_by_dept = df.groupby('Dept')['PhysicalActivityHours'].mean()
plt.bar(mean_activity_by_dept.index, mean_activity_by_dept.values)
plt.xlabel('Department')
plt.ylabel('Mean physical Activity')
plt.title('Mean Physical Activity by dept')
plt.xticks(rotation=90)
plt.show()
```

```python
# Box Plot
plt.boxplot([df[df['Dept'] == dept]['PhysicalActivityHours'] for dept in df['Dept'].unique()])
plt.xlabel('Department')
plt.ylabel('Physical Activity')
plt.title('Physical Activity Distribution by Department')
plt.show()
```

```python
# Violin Plot
import seaborn as sns
import matplotlib.pyplot as plt

# Create a violin plot
sns.violinplot(x='Dept', y='PhysicalActivityHours', data=df)
plt.title('Physical Activity Distribution by Department')
plt.xticks(rotation = 90)
plt.show()
```

```python
# Swarm Plot
import seaborn as sns
import matplotlib.pyplot as plt

# Create a swarm plot
sns.swarmplot(x='Dept', y='PhysicalActivityHours', data=df)
plt.title('Physical Activity vs Department')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a violin plot
sns.violinplot(x='JobLevel', y='Experience', data=df)
plt.title('Experience Distribution by Job Level')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a box plot
plt.boxplot([df[df['JobLevel'] == level]['Experience'] for level in df['JobLevel'].unique()])
plt.xlabel('Job Level')
plt.ylabel('Experience')
plt.title('Experience Distribution by Job Level')
plt.show()
```

```python
df_grouped = df.groupby('Age')['SleepHours'].mean().reset_index()

bins = [20, 25, 30, 35, 40, 45, 50, 55, 60]
labels = ['20-25', '25-30', '30-35', '35-40', '40-45', '45-50', '50-55', '55-60']
df['age_group'] = pd.cut(df['Age'], bins = bins, labels = labels, include_lowest = True)

df_grouped = df.groupby('age_group')['SleepHours'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(df_grouped['age_group'], df_grouped['SleepHours'])
plt.xlabel('Age Group')
plt.ylabel('Average sleep hours')
plt.title('Age vs sleep hours')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
bins = [20, 25, 30, 35, 40, 45, 50, 55, 60]
labels = ['20-25', '25-30', '30-35', '35-40', '40-45', '45-50', '50-55', '55-60']
df['age_group'] = pd.cut(df['Age'], bins = bins, labels = labels, include_lowest = True)

pivot_table = pd.pivot_table(df, index = 'age_group', columns = 'Dept', aggfunc = 'size', fill_value = 0)

plt.figure(figsize=(10, 6))
bottom = np.zeros(len(pivot_table.index))
for i, col in enumerate(pivot_table.columns):
  plt.bar(pivot_table.index, pivot_table[col], bottom = bottom, label = col)
  bottom += pivot_table[col]
plt.xlabel('Age')
plt.ylabel('Number of employees')
plt.title('Age vs department')
plt.legend(pivot_table.columns, title='Department')
plt.show()
```

```python
bins = [20, 25, 30, 35, 40, 45, 50, 55, 60]
labels = ['20-25', '25-30', '30-35', '35-40', '40-45', '45-50', '50-55', '55-60']
df['age_group'] = pd.cut(df['Age'], bins = bins, labels = labels, include_lowest = True)

pivot_table = pd.pivot_table(df, index = 'age_group', columns = 'CommuteMode', aggfunc = 'size', fill_value = 0)

plt.figure(figsize=(10, 6))
bottom = np.zeros(len(pivot_table.index))
for i, col in enumerate(pivot_table.columns):
  plt.bar(pivot_table.index, pivot_table[col], bottom = bottom, label = col)
  bottom += pivot_table[col]
plt.xlabel('Age')
plt.ylabel('Commute Mode')
plt.title('Age vs Commute Mode')
plt.legend(pivot_table.columns, title='Commute Mode')
plt.show()
```

```python
# Commute Distance vs Job level
plt.figure(figsize=(10, 6))
sns.boxplot(x='JobLevel', y='CommuteDistance', data=df)
plt.title('Box plot of commute distance vs job level')
plt.xlabel('Job level')
plt.ylabel('Commute Distance')
plt.show()
```

```python
job_level = ['Senior', "Mid", "Junior", "Lead", "Intern/Fresher"]
edu_level = ["Bachelor", "Master", "High School", "PhD"]

plt.figure(figsize=(10, 8))

contingency_table = pd.crosstab(df['EduLevel'], df['JobLevel'], margins = False)

plt.figure(figsize=(8, 6))
sns.heatmap(contingency_table, annot=False, cmap='YlGnBu',
            xticklabels = edu_level, yticklabels = job_level)

plt.xlabel("Education level")
plt.ylabel("Job level")
plt.title("Education level vs Job level")

plt.show()
```
